```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Fish.csv')
length = df[df['Species'] == 'Perch']['Length1'].values
weight = df[df['Species'] == 'Perch']['Weight'].values

plt.scatter(length, weight)

# x축
x_min = np.min(length)
x_max = np.max(length)

# y축
y_min = weight[np.argmin(length)]
y_max = weight[np.argmax(length)]

plt.plot([x_min, x_max], [y_min, y_max])
# weight[np.argmax(length)]

# 그래프를 여러개 그리기
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

axes[0, 0].scatter([1,2,3,4,5], [10,20,30,40,50])
axes[0, 1].plot([1,2,3,4,5], [10,20,30,40,50])
```
